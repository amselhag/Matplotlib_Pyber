```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

```python
city= pd.read_csv(city_data)
city.head(20)
```

```python
ride= pd.read_csv(ride_data)
ride.head()
```

```python
df=pd.merge(ride, city, on='city', how='outer', indicator='True')
df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
grouped_city=df.groupby(['type','city']).agg({'fare':'mean',
                                      'ride_id':'count',
                                      'driver_count':'mean'})
grouped_city
```

```python
urban_df=grouped_city.loc['Urban']
rural_df=grouped_city.loc['Rural']
suburban_df=grouped_city.loc['Suburban']
```

```python
x_urban= urban_df['ride_id']
y_urban= urban_df['fare']
z_urban = urban_df['driver_count']

x_rural= rural_df['ride_id']
y_rural= rural_df['fare']
z_rural= rural_df['driver_count']

x_suburban= suburban_df['ride_id']
y_suburban= suburban_df['fare']
z_suburban = suburban_df['driver_count']
```

```python
plt.scatter(x_urban, y_urban, s=z_urban*10, alpha=0.5, label="Urban", marker="o", edgecolor='black', color= 'lightcoral', linewidths=2)
plt.scatter(x_suburban, y_suburban, s=z_suburban*10, alpha=0.5, label="Suburban", marker="o", edgecolor='black',color= 'lightskyblue', linewidths=2)
plt.scatter(x_rural, y_rural, s=z_rural*10, alpha=0.5, label="Rural", marker="o", edgecolor='black', color = 'gold', linewidths=2)
plt.legend(title='City Types', labelspacing = 1, markerscale=1, numpoints=1)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title('Pyber Ride Sharing Data (2016)')
plt.grid()
plt.text(45,35, "Note:\nCircle size correlates with driver count per day")
plt.savefig("bubble plot pyber data")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure
```

```python
grouped_type=df.groupby(['type']).agg({'fare':'sum',
                                      'ride_id':'count'})

grouped_type
```

```python
x_fare=grouped_type['fare']
labels=['Rural', 'Suburban', 'Urban']
colors=['gold','lightskyblue','lightcoral']
plt.pie(x_fare, labels=labels, startangle=135, explode=[0,0,0.1], shadow='true', colors = colors, autopct="%1.1f%%")
plt.title('% of Total Fares by City Type')
plt.savefig("Pie chart % of Total Fares by City Type ")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
```

```python
x_rides=grouped_type['ride_id']
plt.pie(x_rides, labels=labels, startangle=135, explode=[0,0,0.1], shadow='true', colors = colors, autopct="%1.1f%%")
plt.title('% of Total Rides by City Type')
plt.savefig("Pie chart % of Total Rides by City Type ")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Build Pie Charts

# Save Figure
```

```python
x_driver=[rural_df['driver_count'].sum(), suburban_df['driver_count'].sum(), urban_df['driver_count'].sum()]
plt.pie(x_driver, labels=labels, startangle=135, explode=[0,0,0.1], shadow='true', colors = colors, autopct="%1.1f%%")
plt.title('% of Total Drivers by City Type')
plt.tight_layout
plt.savefig("Pie chart % of Total Drivers by City Type ")
plt.show()
```

```python
print(f'Average rides per driver by city type:{x_rides/x_driver}')
print(f'Average fare per driver by city type:{x_fare/x_driver}')
```
